### Задание №1

Как относятся друг к другу множество и последовательность? (в ответе использовать слова типа: часть, целое, общее, частное, родитель, дочерний субъект и т.д.)


Последовательностью называется соответствие (бинарное отношение), заданное между множествами натуральных и действительных чисел. Каждому натуральному числу соответствует действительное. Можно расширить определение последовательности, задав соответствие между натуральными числами и любыми объектами (например последовательность матриц). Таким образом, последовательность есть частный случай бинарного отношения между двумя множествами.

### Задание №2

Прочитать высказывания математической логики, построить их отрицания и *установить истинность.

1. $\forall y \in[0 ; 1] : \operatorname{sgn}(y)=1$

Высказывание: для любого у принадлежащего отрезку [0,1] выполняется условие sgn(y)=1.

Высказывание неверно, поскольку sgn(0)=0 по определению функции. 

Отрицание: $\exists y \in[0 ; 1] : \operatorname{sgn}(y)\not=1$

2. $\forall n \in \mathbb{N}>2 : \exists x, y, z \in \mathbb{N} : x^{n}=y^{n}+z^{n}$

Высказывание: Для любого n, принадлежащего множеству натуральных чисел, и больше двух существуют такие x y z, натуральные числа, для которых верно равенство $x^{n}=y^{n}+z^{n}$.

Высказывание неверно, поскольку является отрицанием теоремы Ферма. 

Отрицание: $\forall x,y,z \in \mathbb{N} | \forall n \in \mathbb{N}>2 : x^{n}\not=y^{n}+z^{n}$

3. $\forall x \in \mathbb{R} \exists X \in \mathbb{R} : X>x$

Высказывание: для любого х, принадлежащего множеству действительных чисел, существует большее число. 

Верно, свойство Архимеда.

Отрицание: $\exists x \in \mathbb{R}: \forall X \in \mathbb{R}| X \not =x : X<x$

4. $\forall x \in \mathbb{C} \not\exists y \in \mathbb{C} : x>y \| x<y$

Высказывание: для любого х, принадлежащего множеству комплексных чисел, не существует такого у, принадлежащего множеству комплексных чисел, такого что х больше у или у больше х.

Верно по определению, так как на множестве комплексных чисел не задано отношение порядка больше, меньше.

Отрицание: $\exists y \in \mathbb{C} : \forall x \in \mathbb{C} : x\leq y \| x\geq y$

5. $\forall y \in\left[0 ; \frac{\pi}{2}\right] \exists \varepsilon>0 : \sin y<\sin (y+\varepsilon)$

Высказывание: Для любого у на отрезке от 0 до пи/2 существует такое е, большее 0, такое что $\sin y<\sin (y+\varepsilon)$

Неверно, в точке у = пи/2.

Отрицание:  $\exists y \in\left[0 ; \frac{\pi}{2}\right] : \forall \varepsilon>0 \sin y\geq\sin (y+\varepsilon)$

6. $\forall y \in[0 ; \pi) \exists \varepsilon>0 : \cos y>\cos (y+\varepsilon)$

Высказывание: Для любого у на отрезке 0 до пи существует такое е, большее 0, такое что $\cos y>\cos (y+\varepsilon)$

Верно.

Отрицание:  $\exists y \in[0 ; \pi) : \forall \varepsilon>0 \cos y\leq\cos (y+\varepsilon)$

7. $\exists x : x \notin\{\mathbb{N}, \mathbb{Z}, \mathbb{Q}, \mathbb{R}, \mathbb{C}\}$

Высказывание: существует такой х что он не принадлежит множествам натуральных, целых, рациональных, действительных и комплексных чисел

Верно, к примеру гиперкомплексное число 

Отрицание: $\forall x : x \in\{\mathbb{N}, \mathbb{Z}, \mathbb{Q}, \mathbb{R}, \mathbb{C}\}$

```python
import itertools
```

```python
a = {2,4,6,10,12,5}
b = {3,4,5,13,23,10,32}
c = {2}
```

```python
def inter(a,b):
    return a.intersection(b)
def union(a,b):
    return a.union(b)
def difference(a,b):
    return a.difference(b)
def symmetric_difference(a,b):
    return a.symmetric_difference(b)
def multiplication(a,b):
    return { _ for _ in itertools.product(a,b)}
```
